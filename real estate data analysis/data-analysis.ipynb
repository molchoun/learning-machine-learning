```python
import pandas as pd

df = pd.read_csv('ad-data-apt.csv')
df.head()
```

```python
df['Price'] = df['Price'].str.replace(',', '').str.replace('$', '').str.replace('֏', '').str.replace('€', '').str.replace('руб.', '')
df['Price']
```

```python
df['Price'] = df['Price'].apply (pd.to_numeric, errors='coerce')
df['Price'] = df['Price'].dropna()
df['Price']
```
